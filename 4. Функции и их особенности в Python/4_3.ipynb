```python
def recursive_sum(*numbers) -> int:
    if not numbers:
        return 0
    else:
        return numbers[-1] + recursive_sum(*numbers[:-1])
```

**B. Рекурсивный сумматор цифр**

image.png

```python
def recursive_digit_sum(number) -> int:
    if number == 0:
        return 0
    else:
        return number % 10 + recursive_digit_sum(number // 10)
```

```python
def result_accumulator(f):
    accumulator = dict()

    def decorated(*args, method='accumulate'):
        if method == 'accumulate':
            accumulator[f] = accumulator.get(f, []) + [f(*args)]
        elif method == 'drop':
            answer = accumulator[f] + [f(*args)]
            accumulator[f] = []
            return answer
    return decorated
```

**F. Сортировка слиянием**

image.png

```python
def merge_sort(list_of_numbers):
    if len(list_of_numbers) <= 1:
        return list_of_numbers

    mid = len(list_of_numbers) // 2

    left = merge_sort(list_of_numbers[:mid])
    right = merge_sort(list_of_numbers[mid:])

    left_index = 0
    right_index = 0

    sorted_list = []
    while left_index < len(left) and right_index < len(right):
        if left[left_index] <= right[right_index]:
            sorted_list += [left[left_index]]
            left_index += 1
        else:
            sorted_list += [right[right_index]]
            right_index += 1

    while left_index < len(left):
        sorted_list += [left[left_index]]
        left_index += 1

    while right_index < len(right):
        sorted_list += [right[right_index]]
        right_index += 1

    return sorted_list
```

```python
def same_type(f):
    def decorated(*args):
        types = set()
        for arg in args:
            types.add(type(arg))
            if len(types) > 1:
                print("Обнаружены различные типы данных")
                return

        return f(*args)
    return decorated
```

**H. Генератор Фибоначчи**


image.png


image.png

```python
def fibonacci(n):
    n_1, n_2 = 0, 1
    for i in range(n):
        yield n_1
        n_1, n_2 = n_2, n_1 + n_2
```

```python
def cycle(iterable):
    while iterable:
        for element in iterable:
            yield element
```

```python
def make_linear(not_linear_list):
    linear = []
    for element in not_linear_list:
        if isinstance(element, list):
            linear += make_linear(element)
        else:
            linear += [element]

    return linear
```
