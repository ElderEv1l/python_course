```python
def secret_replace(message, **kwargs):
    secret_message = ''

    # счётчик кол-ва замен буквы в уже пройденной строке
    counter = {letter: 0 for letter in kwargs}
    for letter in message:
        if letter in kwargs:
            # если для буквы есть замена - заменяем на ту что есть в словаре по индексу счетчика
            secret_message += kwargs[letter][counter[letter] % len(kwargs[letter])]
            counter[letter] += 1
        else:
            secret_message += letter

    return secret_message
```
