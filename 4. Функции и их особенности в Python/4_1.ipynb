```python
def gcd(a, b):
    while a != 0 and b != 0:
        if a > b:
            a %= b
        else:
            b %= a
    return a + b
```

```python
def month(month_number, language):
    if language == 'ru':
        months = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь',
                  'Ноябрь', 'Декабрь']
        return months[month_number - 1]
    if language == 'en':
        months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October',
                  'November', 'December']
        return months[month_number - 1]
```

```python
def is_palindrome(a):
    if isinstance(a, int):
        a = str(a)
    a = list(a)
    for char in a:
        if char != a[-1]:
            return False
        a.pop(-1)
    return True
```

```python
from math import sqrt


def is_prime(a):
    for i in range(2, int(sqrt(a))):
        if a % i == 0:
            return False
    return True
```

```python
def merge(a: tuple, b: tuple):
    merged = []
    while len(a) != 0 and len(b) != 0:
        if a[0] <= b[0]:
            merged += a[:1]
            a = a[1:]
        else:
            merged += b[:1]
            b = b[1:]

    if len(a) == 0:
        merged += b
    else:
        merged += a
    return *merged,
```
