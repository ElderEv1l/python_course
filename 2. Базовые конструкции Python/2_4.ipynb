**H. Максимальная сумма**

image.png

```python
winner = ""
max = 0

for i in range(int(input())):
    name = input()
    a = int(input())
    tmp = a
    sum = 0
    while tmp != 0:
        sum += tmp % 10
        tmp //= 10

    if max <= sum:
        winner = name
        max = sum

print(winner)
```

**P. Редизайн таблицы умножения**

image.png


image.png

```python
n = int(input())
width = int(input())

w = (width) * n + n - 1

for i in range(1, n + 1):
    for j in range(1, n + 1):
        print(f"{i * j:^{width}}", end='')
        if j != n:
            print("|", end="")
    print()
    if i != n:
        print(f'{"" :-<{w}}')
```

**R. Новогоднее настроение 2.0**


image.png


image.png


image.png

```python
# тестовый забег для вычисления width последней строки
last = 1
step = 2

for i in range(1, (n := int(input())) + 1):
    if i == last:
        last += step
        step += 1

width = 0
for i in range(n, n - step + 2, -1):
    width += len(str(i))
    if i != n - step + 3:
        width += 1


# итоговый забег
last = 1
step = 2

out = ''
for i in range(1, n + 1):
    out += str(i)
    if i == last or i == n:
        print(f"{out:^{width}}")
        out = ''
        last += step
        step += 1
    else:
        out += ' '
```

**T. Математическая выгода**


image.png


image.png

```python
a = int(input())
max_sum = 0
profit_system = 0

for i in range(2, 11):
    sum_of_digits = 0
    tmp = a
    while tmp != 0:
        sum_of_digits += tmp % i
        tmp //= i
    if sum_of_digits > max_sum:
        max_sum = sum_of_digits
        profit_system = i

print(profit_system)
```
