**N. Властелин Чисел: Две Башни**

image.png

```python
number = int(input())

min = number % 10
max = number % 10

if min > number // 10 % 10:
    min = number // 10 % 10
if min > number // 100:
    min = number // 100

if max < number // 10 % 10:
    max = number // 10 % 10
if max < number // 100:
    max = number // 100

left = number % 10 + number // 10 % 10 + number // 100 - min - max

if min != 0:
    print(f"{min}{left} {max}{left}")
else:
    print(f"{left}{min} {max}{left}")
```

```python
x = float(input())
y = float(input())

# y = 1/4 x^2 + 1/2 x - 35/4
# for [-7;4] y = 5/3 x + 35/3
# for [4;0] y = 5
# for [0;5]

if x ** 2 + y ** 2 > 100:
    print("Вы вышли в море и рискуете быть съеденным акулой!")
elif ((-10 <= x < -7) or (x == -7 and y != 0)) or ((5 < x <= 10) or (x == 5 and y != 0)):
    print("Зона безопасна. Продолжайте работу.")
elif (y < (1 / 4) * x ** 2 + (1 / 2) * x - (35 / 4)):
    print("Зона безопасна. Продолжайте работу.")
elif -7 <= x <= -4:
    if y > (5 / 3) * x + (35 / 3):
        print("Зона безопасна. Продолжайте работу.")
    else:
        print("Опасность! Покиньте зону как можно скорее!")
elif -4 <= x <= 0:
    if y > 5:
        print("Зона безопасна. Продолжайте работу.")
    else:
        print("Опасность! Покиньте зону как можно скорее!")
elif 0 <= x <= 5:
    if (x ** 2 + y ** 2 > 25) and (y > 0):
        print("Зона безопасна. Продолжайте работу.")
    else:
        print("Опасность! Покиньте зону как можно скорее!")
```
