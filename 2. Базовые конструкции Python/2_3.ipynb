**J. Маршрут построен**

```python
x = 0
y = 0

while (tmp := input()) != 'СТОП':
    match tmp:
        case 'СЕВЕР':
            x += int(input())
        case 'ЮГ':
            x -= int(input())
        case 'ЗАПАД':
            y -= int(input())
        case 'ВОСТОК':
            y += int(input())

print(f"{x}\n{y}")
```

```python
from math import sqrt
n = int(input())
for i in range(2, int(sqrt(n))):
    while n % i == 0:
        print(i, end='')
        n //= i
        if n != 1:
            print(" * ", end='')
if n != 1:
    print(n)
```

**S. Игра в «Угадайку»**


image.png


image.png

```python
left = 1
right = 1000
middle = (left + right) // 2
print(middle)

while (tmp := input()) != "Угадал!":
    if tmp == "Больше":
        left = middle + 1
    else:
        right = middle - 1
    print(middle := (left + right) // 2)
```
