**T. Мухи отдельно, котлеты отдельно**


image.png


image.png

```python
weight = int(input())
price = int(input())
price_first = int(input())
price_second = int(input())

# решим аналитически уравнение
# (x*k1+y*k2)/(x+y)=m
# x*k1+y*k2=x*m+y*m
# x(k1-m)+y(k2-m)=0
# x(k1-m)=(N-x)(m-k2)
# x*k1-x*m = n*(m-k2) +x*(k2-m)
# x*k1-x*m-x*k2+x*m=n*(m-k2)
# x=n*(m-k2)/(k1-k2)

weight_first = int(weight * (price - price_second) / (price_first - price_second))
print(f"{weight_first} {weight - weight_first}")
```
