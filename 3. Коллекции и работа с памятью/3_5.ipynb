**K. Файловая статистика 2.0**

```python
import json

with open(input(), encoding="UTF-8") as file_in:
    inp = file_in.read().split()

counter = 0
counter_of_positive = 0
minimum = int(inp[0])
maximum = int(inp[0])
total = 0

for number in inp:
    counter += 1
    if int(number) > 0:
        counter_of_positive += 1
    if int(number) < minimum:
        minimum = int(number)
    if int(number) > maximum:
        maximum = int(number)
    total += int(number)

result = {
    'count': counter,
    'positive_count': counter_of_positive,
    'min': minimum,
    'max': maximum,
    'sum': total,
    'average': round(total / counter, 2)}

with open(input(), "w", encoding="UTF-8") as file_out:
    json.dump(result, file_out, indent=2)
```

**L. Разделяй и властвуй**


image.png


image.png


image.png

```python
with open(input(), encoding="UTF-8") as file_in:
    inp = file_in.readlines()

with open(input(), "w", encoding="UTF-8") as file_out_even, open(input(), "w", encoding="UTF-8") as file_out_odd, open(
        input(), "w", encoding="UTF-8") as file_out_eq:
    for line in inp:
        for number in line.split():
            even = 0
            odd = 0
            a = int(number)
            while a != 0:
                if (a % 10) % 2 == 0:
                    even += 1
                else:
                    odd += 1
                a //= 10
            if even > odd:
                print(int(number), end=' ', file=file_out_even)
            elif even < odd:
                print(int(number), end=' ', file=file_out_odd)
            else:
                print(int(number), end=' ', file=file_out_eq)

        print(file=file_out_even)
        print(file=file_out_odd)
        print(file=file_out_eq)
```

**N. Слияние данных**


image.png


image.png


image.png

```python
import json

with open(filename := input(), encoding="UTF-8") as file_in:
    old_records = json.load(file_in)

with open(input(), encoding="UTF-8") as file_in:
    updates = json.load(file_in)

result = dict()
for record in old_records:
    result[record.pop('name')] = record

for record in updates:
    name = record.pop('name')
    if name not in result:
        result[name] = record
    else:
        for key, value in record.items():
            if key not in result[name] or value > result[name][key]:
                result[name][key] = value

with open(filename, "w", encoding="UTF-8") as file_out:
    json.dump(result, file_out, ensure_ascii=False, indent=2)
```

```python
with open('secret.txt', encoding="UTF-8") as file_in:
    message = file_in.read().splitlines()

for i in range(len(message)):
    new_line = ''
    for letter in message[i]:
        if ord(letter) >= 128:
            letter = chr(int(ord(letter) & 0xFF))
        new_line += letter
    message[i] = new_line

print('\n'.join(message))
```

**S. Это будет наш секрет**


image.png


image.png


image.png

```python
import string

shift = int(input())
with open('public.txt', encoding="UTF-8") as file_in:
    message = file_in.read()

alphabet = string.ascii_uppercase
private_message = ''
for letter in message:
    if letter.upper() not in alphabet:
        private_message += letter
    elif letter.isupper():
        private_message += alphabet[(alphabet.find(letter.upper()) + shift) % 26].capitalize()
    else:
        private_message += alphabet[(alphabet.find(letter.upper()) + shift) % 26].lower()

with open('private.txt', "w", encoding="UTF-8") as file_out:
    print(private_message, file=file_out)
```
