**O. Двоичная статистика!**

image.png

```python
list = []

for i in (input().split()):
    a = int(i)
    stat = {
        "digits": 0,
        "units": 0,
        "zeros": 0}

    while a > 0:
        stat['digits'] += 1
        if a % 2 == 0:
            stat['zeros'] += 1
        else:
            stat['units'] += 1
        a = a // 2
    list.append(stat.copy())

print(list)
```

**Q. Друзья друзей**


image.png


image.png


image.png

```python
relations = dict()

while relation := input():
    relations[relation.split()[0]] = relations.get(relation.split()[0], []) + [relation.split()[1]]
    relations[relation.split()[1]] = relations.get(relation.split()[1], []) + [relation.split()[0]]

second_hand = set()
for key in sorted(relations.keys()):
    second_hand.clear()
    for person in relations[key]:
        second_hand |= set(relations[person])
    second_hand -= set(relations[key])
    second_hand -= {key}
    print(f"{key}: {', '.join(sorted(second_hand))}")
```

**S. Частная собственность**


image.png


image.png

```python
toys = dict()

for i in range(int(input())):
    inp = input()
    for toy in inp[inp.find(' ') + 1:].split(', '):
        toys[inp[:inp.find(':')]] = toys.get(inp[:inp.find(':')], set()) | {toy}

result = set()
for i in toys.keys():
    tmp = toys[i].copy()
    for j in toys.keys():
        if i != j:
            tmp -= toys[j]
    result |= tmp
print('\n'.join(sorted(result)))
```

```python
import math

result = dict()

numbers = sorted(set(map(int, input().split('; '))))

for i in range(len(numbers)):
    for j in range(i, len(numbers)):
        if math.gcd(numbers[i], numbers[j]) == 1:
            result[numbers[i]] = result.get(numbers[i], []) + [numbers[j]]
            result[numbers[j]] = result.get(numbers[j], []) + [numbers[i]]
    if numbers[i] in result:
        print(f'{numbers[i]} - {", ".join(map(str, result[numbers[i]]))}')
```
