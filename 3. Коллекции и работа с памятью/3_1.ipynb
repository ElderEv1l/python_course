**P. Анонс новости 2.0**


image.png


image.png


image.png

```python
max_length = int(input())
header = []

for i in range(int(input())):
    header.append(input())

current_length = 0
for i in range(len(header)):
    current_length += len(header[i])

    if current_length > max_length:
        print(header[i][:max_length - current_length - 3] + '...')
        break
    elif max_length - current_length <= 3:
        if i == len(header) - 1:
            print(header[i])
        else:
            print(header[i][:max_length - current_length - 3] + '...')
        break
    elif max_length - current_length == 3:
        print(header[i] + '...')
        break
    else:
        print(header[i])
```

```python
a = input()

count = 1
for i in range(len(a) - 1):
    if a[i] == a[i + 1]:
        count += 1
        if i == len(a) - 2:
            print(f"{a[i]} {count}")
    else:
        print(f"{a[i]} {count}")
        count = 1
        if i == len(a) - 2:
            print(f"{a[i + 1]} {count}")
```

**S. Польский калькулятор**


image.png


image.png


image.png

```python
line = input().split()
digits = []

for c in line:
    match c:
        case '+':
            x = digits.pop()
            y = digits.pop()
            digits.append(x + y)
        case '-':
            x = digits.pop()
            y = digits.pop()
            digits.append(y - x)
        case '*':
            x = digits.pop()
            y = digits.pop()
            digits.append(x * y)
        case _:
            digits.append(int(c))

print(digits[0])
```

**T. Польский калькулятор — 2**


image.png


image.png


image.png

```python
import math

line = input().split()
digits = []

for c in line:
    match c:
        case '+':
            x = digits.pop()
            y = digits.pop()
            digits.append(x + y)
        case '-':
            x = digits.pop()
            y = digits.pop()
            digits.append(y - x)
        case '*':
            x = digits.pop()
            y = digits.pop()
            digits.append(x * y)
        case '/':
            x = digits.pop()
            y = digits.pop()
            digits.append(y // x)
        case '~':
            x = digits.pop()
            digits.append(-x)
        case '!':
            x = digits.pop()
            digits.append(math.factorial(x))
        case '#':
            x = digits.pop()
            digits.append(x)
            digits.append(x)
        case '@':
            x = digits.pop()
            y = digits.pop()
            z = digits.pop()
            digits.append(y)
            digits.append(x)
            digits.append(z)
        case _:
            digits.append(int(c))

print(digits[0])
```
