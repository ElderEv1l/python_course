```python
from itertools import product

operators = {
    'not': 'not',
    'and': 'and',
    'or': 'or',
    '^': '!=',
    '->': '<=',
    '~': '==',
}

priority = {
    'not': 0,
    'and': 1,
    'or': 2,
    '^': 3,
    '->': 4,
    '~': 5,
    '(': 6,
}


def make_postfix_expression(exp, used_variables):
    stack = []
    postfix = []
    for i in exp.split():
        if i in used_variables:
            postfix.append(i)
        elif i == '(':
            stack.append(i)
        elif i == ')':
            while stack[-1] != '(':
                postfix.append(operators[stack.pop()])
            stack.pop()
        elif i in operators.keys():
            while len(stack) and priority[i] >= priority[stack[-1]]:
                postfix.append(operators[stack.pop()])
            stack.append(i)

    for _ in range(len(stack)):
        postfix.append(operators[stack.pop()])

    return postfix


def get_result_of_postfix_expression(postfix_exp, used_variables):
    stack = []
    for i in range(len(postfix_exp)):
        if postfix_exp[i] in used_variables.keys():
            stack.append(used_variables[postfix_exp[i]])
        else:
            if postfix_exp[i] == 'not':
                stack.append(not stack.pop())
            else:
                y, x = stack.pop(), stack.pop()
                stack.append(eval(f'{x} {postfix_exp[i]} {y}'))
    return int(stack.pop())


expression = input()
variables = [variable for variable in 'ABCDE' if variable in expression]

statement = expression.replace('(', '( ').replace(')', ' )')
postfix_expression = make_postfix_expression(statement, variables)

values = product(range(2), repeat=len(variables))
print(' '.join(variables) + ' F')
for kit in values:
    dic = dict(zip(variables, kit))
    for i in range(len(kit)):
        print(kit[i], end=' ')
    print(get_result_of_postfix_expression(postfix_expression, dic))
```
