```python
class Programmer:
    def __init__(self, name, post):
        self.name = name
        self.post = post
        self.hours = 0
        self.accumulated_salary = 0
        match self.post:
            case 'Junior':
                self.salary_per_hour = 10
            case 'Middle':
                self.salary_per_hour = 15
            case 'Senior':
                self.salary_per_hour = 20

    # в данном методе мы добавляем часы в отработанное время и прибавляем накопленную зарплату
    def work(self, hours):
        self.hours += hours
        self.accumulated_salary += self.salary_per_hour * hours

    def rise(self):
        match self.post:
            case 'Junior':
                self.post = 'Middle'
                self.salary_per_hour = 15
            case 'Middle':
                self.post = 'Senior'
                self.salary_per_hour = 20
            case 'Senior':
                self.salary_per_hour += 1

    def info(self):
        return f"{self.name} {self.hours}ч. {self.accumulated_salary}тгр."
```

```python
class Rectangle:
    def __init__(self, first_point, second_point):
        self.left_upper_corner = min(first_point[0], second_point[0]), max(first_point[1],
                                                                           second_point[1])
        self.right_lower_corner = max(first_point[0], second_point[0]), min(first_point[1],
                                                                            second_point[1])
        self.width = abs(first_point[0] - second_point[0])
        self.height = abs(first_point[1] - second_point[1])

    def perimeter(self):
        return round(2 * (self.width + self.height), 2)

    def area(self):
        return round(self.width * self.height, 2)

    def get_pos(self):
        return round(self.left_upper_corner[0], 2), round(self.left_upper_corner[1], 2)

    def get_size(self):
        return round(self.width, 2), round(self.height, 2)

    def move(self, dx, dy):
        self.left_upper_corner = self.left_upper_corner[0] + dx, self.left_upper_corner[1] + dy
        self.right_lower_corner = self.right_lower_corner[0] + dx, self.right_lower_corner[1] + dy

    def resize(self, width, height):
        self.width = width
        self.height = height
        self.right_lower_corner = self.left_upper_corner[0] + width, self.left_upper_corner[1] - height

    def turn(self):
        # находим центр прямоугольника, чтобы сделать вокруг него поворот
        middle_point = self.left_upper_corner[0] + round(self.width / 2, 2), self.left_upper_corner[1] - round(
            self.height / 2, 2)

        # меняем местами высоту и ширину объекта
        self.width, self.height = self.height, self.width

        # находим новые точки прямоугольника
        self.left_upper_corner = middle_point[0] - round(self.width / 2, 2), middle_point[1] + round(self.height / 2, 2)
        self.right_lower_corner = middle_point[0] + round(self.width / 2, 2), middle_point[1] - round(self.height / 2,
                                                                                                      2)

    def scale(self, factor):
        # аналогично повороту находим центр прямоугольника, чтобы изменить размер
        middle_point = self.left_upper_corner[0] + round(self.width / 2, 2), self.left_upper_corner[1] - round(
            self.height / 2, 2)

        # меняем местами высоту и ширину объекта
        self.width, self.height = round(self.width * factor, 2), round(self.height * factor, 2)

        # находим новые точки прямоугольника
        self.left_upper_corner = middle_point[0] - round(self.width / 2, 2), middle_point[1] + round(self.height / 2, 2)
        self.right_lower_corner = middle_point[0] + round(self.width / 2, 2), middle_point[1] - round(self.height / 2,
                                                                                                      2)
```

```python
class Checkers:
    def __init__(self):
        self.field = dict()
        for index_of_the_letter, letter in enumerate('ABCDEFGH'):
            for index_of_the_number, number in enumerate('12345678'):
                #
                if number in '45':
                    self.field[letter + number] = Cell('X')
                    continue

                if number in '123678':
                    if (index_of_the_letter + index_of_the_number) % 2 == 0:
                        if number in '123':
                            self.field[letter + number] = Cell('W')
                        else:
                            self.field[letter + number] = Cell('B')
                    else:
                        self.field[letter + number] = Cell('X')
                    continue

    def move(self, f, t):
        self.field[t] = self.field[f]
        self.field[f] = Cell('X')

    def get_cell(self, p):
        return self.field[p]


class Cell:
    def __init__(self, cell_state):
        self.state = cell_state

    def status(self):
        return self.state
```

```python
class Queue:
    def __init__(self):
        self.queue = []

    def push(self, item):
        self.queue += [item]

    def pop(self):
        return self.queue.pop(0)

    def is_empty(self):
        return len(self.queue) == 0
```
