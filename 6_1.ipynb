**H. Числовая змейка 3.0**

image.png

```python
import numpy as np


def snake(m, n, direction='H'):
    result = np.array([i for i in range(1, m * n + 1)], dtype=np.int16)
    if direction == 'H':
        result.shape = (n, m)
        for i in range(len(result)):
            if i % 2 == 1:
                result[i] = np.flip(result[i])
    elif direction == 'V':
        result.shape = (m, n)
        for i in range(len(result)):
            if i % 2 == 1:
                result[i] = np.flip(result[i])
        result = np.transpose(result)

    return result
```

```python
import numpy as np


def stairs(vector):
    result = np.array([np.roll(vector, i) for i in range(len(vector))])
    return result
```
